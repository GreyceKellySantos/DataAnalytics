```python
# Step 1. Load the dataset into Jupyter Notebook.
# https://www.kaggle.com/datasets/valakhorasani/gym-members-exercise-dataset

# Step 2. Look for the shape of the data.

import pandas as pd

dataset = pd.read_csv('gym_tracking.csv')
dataset
```

```python
# Step 3.Describe the data quantitatively.

dataset.describe()
```

```python
# Step 4.Perform thorough data cleaning of the dataset.
cleanedData = dataset.dropna()
cleanedData
```

```python
# Step 5.Select 7 variables of interest to work with.
VariableOfInterest = dataset [["Weight (kg)", "Max_BPM","Avg_BPM", "Resting_BPM", "Age", "Fat_Percentage", "BMI"]]
VariableOfInterest
```

```python
correlation_matrix = VariableOfInterest.corr()
correlation_matrix
```

```python
# Step 6.Plot boxplots to identify whether we have outliers.

import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (28,20))            
count = 0                                       

for variable in VariableOfInterest:            
    count += 1
    plt.subplot (3,3, count)
    ax = sns.boxplot (y = variable, data = dataset) 
plt.show()                                      
```

```python
# Step 7.Remove the outliers and plot new boxplots to see the results.

plt.figure(figsize=(10,4))
VariableOfInterest.boxplot()

plt.title("Box Plot | Variables of Interest")
plt.ylabel("Values")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 8.Describe the results seen in the boxplots
Initial_Null_Value = dataset.isnull().sum().sum()

import numpy as np

dataset.loc[dataset["Weight (kg)"]>125,"Weight (kg)"] = np.nan
dataset.loc[dataset["Weight (kg)"]<0,"Weight (kg)"] = np.nan

dataset.loc[dataset["BMI"]>40,"BMI"] = np.nan
dataset.loc[dataset["BMI"]<0,"BMI"] = np.nan

Final_Null_Value = dataset.isnull().sum().sum()
```

```python
print("These are the observations removed:" + str(Final_Null_Value - Initial_Null_Value))

print("There are the observations that comform the outliers:" 
      + str(round((Final_Null_Value - Initial_Null_Value) *100/dataset.shape[0],2))+"%")
```
